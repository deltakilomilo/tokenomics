```python
import pandas as pd
dfs = []
dfs.append((pd.read_csv('../data/acs-usd-max.csv'), 'acs'))
dfs.append((pd.read_csv('../data/alpaca-usd-max.csv'), 'alpaca'))
dfs.append((pd.read_csv('../data/alpha-usd-max.csv'), 'alpha'))
dfs.append((pd.read_csv('../data/auto-usd-max.csv'), 'auto'))
dfs.append((pd.read_csv('../data/bfi-usd-max.csv'), 'bfi'))
dfs.append((pd.read_csv('../data/bifi-usd-max.csv'), 'bifi'))
dfs.append((pd.read_csv('../data/bunny-usd-max.csv'), 'bunny'))
dfs.append((pd.read_csv('../data/farm-usd-max.csv'), 'harvest'))
dfs.append((pd.read_csv('../data/idle-usd-max.csv'), 'idle'))
dfs.append((pd.read_csv('../data/pickle-usd-max.csv'), 'pickle'))
dfs.append((pd.read_csv('../data/value-usd-max.csv'), 'value'))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Price action since inception
for i, obj in enumerate(dfs):
    plt.figure()
    df = obj[0]
    name = obj[1]
    plt.plot(df['snapped_at'], df['price'])
    plt.title(name)
    plt.text(1, 1, f'Start date: {df["snapped_at"][0]}')
plt.show()
```

```python
# Price action absolute
plt.figure()
df_btc = pd.read_csv('../data/btc-usd-max.csv')
for i, obj in enumerate(dfs):
    df = obj[0]
    name = obj[1]
    plt.plot(df['snapped_at'], df['price'], 'r')
    plt.title(name)
    # plt.text(1, 1, f'Start date: {df["snapped_at"][0]}')
    # Price action bitcoin
    plt.plot(df_btc['snapped_at'], df_btc['price']/10.0, 'g')
    plt.title('BTC')
plt.show()
```
